```python
#Question 3
"""
Consider the given breast cancer dataset from the given google drive link, the last column has the target variable (2 classes: "B" for benign and "M" for malignant) and the remaining columns have the features.
Implement the decision tree algorithm from scratch. You can use the Gini index to decide on the splitting attribute on the breast cancer dataset. Make use of only NumPy, pandas. Calculate the accuracy of the decision tree.
"""
```

```python
breastCancer = pd.read_csv("wdbc.csv", header = None)
training_data, testing_data = train_test_split(breastCancer, test_size = 0.3, random_state = 30)

training_data = training_data.to_numpy().tolist()
testing_data = testing_data.to_numpy().tolist()
```

```python
def distinct(rows):
    count = {}
    for row in rows:
        label = row[1]
        if label not in count:
            count[label] = 0
        count[label] += 1
    return count  
```

```python
class Node:
    
    def __init__(self, true_child, false_child, question, rows):
        self.true_child = true_child
        self.false_child = false_child
        self.question = question
        self.rows = rows
        
def gini(rows):
    counts = distinct(rows)
    gini_val = 1
    for i in counts:
        ans = ((counts[i]) / (len(rows))) ** 2
        gini_val = gini_val - ans
    return gini_val
    
def info_gain(ginival, true_child, false_child):
    p = len(true_child) / (len(true_child) + len(false_child))
    info_g = ginival - ((p * gini(true_child)) + ((1 - p) * gini(false_child)))
    return info_g
    
def best_split(rows):
    best_gain = 0
    best_question = None
    current_gini = gini(rows)
    n = len(rows[0])  
    for col in range(n):
        if (col == 1):
            continue
        options = [row[col] for row in rows]
        for val in options:
            question = Classifier(col, val)
            true_child, false_child = splitset(rows, question)
            if (len(true_child) == 0 | len(false_child) == 0): #lite
                continue 
            gain = info_gain(current_gini, true_child, false_child)
            best_gain, best_question = [gain, question] if gain > best_gain else [best_gain, best_question]

    return best_gain, best_question
```

```python
class Leaf_node:
    def __init__(self, rows):
        self.dataset = distinct(rows)
```

```python
class Classifier:
    
    def __init__(self, column, value):
        self.column = column
        self.value = value
        
    def match(self, sample):
        val = sample[self.column]
        return val >= self.value
```

```python
def splitset(dataset, classification):
    matchset, nomatchset = [], []
    for row in dataset:
        if classification.match(row):
            matchset.append(row)
        else:
            nomatchset.append(row)
    return matchset, nomatchset
```

```python
def build(rows):
    gain, question = best_split(rows)
    if gain == 0:
        return Leaf_node(rows)
    
    matchset, nonmatchset = splitset(rows, question)
    true_child = build(matchset)
    false_child = build(nonmatchset)
    return Node(true_child, false_child, question, rows)
```

```python
tree = build(training_data)
```

```python
def decision(inp, node):
    if type(node) is Leaf_node:
        return node.dataset
    
    if node.question.match(inp):
        return decision(inp, node.true_child)
    else:
        return decision(inp, node.false_child)
    
correct_count = 0
for i in testing_data:
    key = list(decision(i, tree).keys())
    if (i[1] == key[0]):
        correct_count += 1
print("Accuracy:", str((correct_count / len(testing_data) * 100)) + "%")
```
